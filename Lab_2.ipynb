```python
# Exercise #11
import numpy as np
import matplotlib.pyplot as plt

A = np.zeros((500, 500), dtype=np.uint8)

A[100:120, 100:400] = 255
A[120:400, 240:260] = 255

plt.figure(8)
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Exercise #12
A = np.zeros((500, 500), dtype=np.uint8)

A[1::2, 1::2] = 255

plt.figure(figsize=(6, 6))
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
# Exercise #13
# Create a 500x500 array filled with zeros
A = np.zeros((500, 500), dtype=np.uint8)

for i in range(500):
    for j in range(500):
        if 2 * (i // 2) != i and 2 * (j // 2) != j:
            A[i, j] = 255
plt.figure(10)
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Exercise #14
import cv2

A = cv2.imread('/Users/klnimri/Downloads/Spine.png', cv2.IMREAD_GRAYSCALE)

A_min = np.min(A)
A_max = np.max(A)
print([A_min, A_max])

B = ((A.astype(float) - A_min) / (A_max - A_min)) * 255.0

C = B - A.astype(float)

plt.figure(11, figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
plt.title("A")
plt.subplot(1, 3, 2)
plt.imshow(B, cmap='gray', vmin=0, vmax=255)
plt.title("B")
plt.subplot(1, 3, 3)
plt.imshow(C, cmap='gray', vmin=-255, vmax=255)
plt.title("C")
plt.show()
```
